```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('merged_HiSeqV2')
```

```python
unique_values = df['histological_type'].unique()
unique_values
```

```python
count_adenosquamous = (df['histological_type'] == 'Adenosquamous').sum()
count_adenosquamous
```

```python
df.shape
```

```python
mask = df['histological_type'] == 'Adenosquamous'

# Drop the rows where the mask is True (i.e., 'histological_type' is 'Adenosquamous')
df.drop(df[mask].index, inplace=True)
```

```python
unique_values = df['histological_type'].unique()
unique_values
```

```python
mapping = {
    'Endocervical Type of Adenocarcinoma': 'Adenocarcinoma',
    'Endometrioid Adenocarcinoma of Endocervix': 'Adenocarcinoma',
    'Endocervical Adenocarcinoma of the Usual Type': 'Adenocarcinoma',
    'Mucinous Adenocarcinoma of Endocervical Type': 'Adenocarcinoma'
}

# Use the replace method to change the values in the 'Histologic' column
df['histological_type'] = df['histological_type'].replace(mapping)
```

```python
unique_values = df['histological_type'].unique()
unique_values
```

```python
missing_values = df['histological_type'].isna().sum()

if missing_values > 0:
    print(f"There are {missing_values} missing values in the 'Histological_type' column.")
else:
    print("There are no missing values in the 'Histological_type' column.")
```

```python
df.dropna(subset=['histological_type'], inplace=True)
```
