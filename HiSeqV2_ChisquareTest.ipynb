```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('After_subtypes_simplified_HiSeqV2')
```

```python
df
```

```python
# Step 2: Remove the first column (index 0)
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df
```

```python
X = df.iloc[:,0:-1]
y = df["histological_type"]
```

```python
## Perform chi2 test
### chi2 returns 2 values
### Fscore and the pvalue
from sklearn.feature_selection import chi2
f_p_values=chi2(X,y)
```

```python
f_p_values
```

```python
import pandas as pd
p_values=pd.Series(f_p_values[1])
p_values.index=X.columns
p_values
```

```python
sorted_p_values = p_values.sort_values(ascending=False)
sorted_p_values
```

```python
import matplotlib.pyplot as plt
# Create a bar plot of the F-scores
plt.figure(figsize=(12, 6))
sorted_p_values.plot(kind='bar')
plt.title('F-score Values for Features (Sorted)')
plt.xlabel('Features')
plt.ylabel('F-score')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()
plt.show()
```
