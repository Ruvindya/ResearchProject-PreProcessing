```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('normalized_GeneExpressionHiSeq')
```

```python
df
```

```python
clinical = pd.read_csv('phenotype',delimiter='\t')
```

```python
clinical
```

```python
# Extract the specific column from dataset1
column_to_extract = clinical[['sampleID', 'histological_type'] ]
column_to_extract
```

```python
merged_df = pd.merge(df,column_to_extract, left_on='index', right_on='sampleID')
```

```python
merged_df
```

```python
merged_df[['index', 'sampleID','histological_type']]
```

```python
are_columns_same = (merged_df['index'] == merged_df['sampleID']).all()

if are_columns_same:
    print("The index and 'SampleId' column are the same in the merged dataset.")
else:
    print("The index and 'SampleId' column are not the same in the merged dataset.")
```

```python
# Drop the 'SampleId' column
merged_df.drop(columns='sampleID', inplace=True)
```
